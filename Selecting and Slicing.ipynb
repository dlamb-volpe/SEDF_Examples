### Select Using Geometry

Selecting points that intersect the polygon. You need to use the points layer select and pass the polygon to the function.

```python
sdf_points.spatial.select(sdf_polygon)
```

Slice using pandas .loc.

```python
sdf_points.spatial.select(sdf_polygon.loc[2:4])
```

```python
sdf_points.spatial.select(sdf_polygon[pd.isnull(sdf_polygon["CENTROID_X"])])
```

### Using a slice and spatial.select causes an error

```python
sdf_points[sdf_points["OBJECTID"]%2==0].spatial.select(sdf_polygon)
```

### Solution is to copy and reset index

```python
temp = sdf_points[sdf_points["OBJECTID"]%2==0].copy().reset_index()
temp.spatial.select(sdf_polygon)
```
